```python
import random
import numpy as np
import time
import matplotlib.pyplot as plt
def large_arrays_choice_numpy(N):
    a = np.random.choice(range(101),size=N)
    b = np.random.choice(range(101),size=N)
    return a,b

def add_large_arrays_choice(N):
    a = random.choices(range(101),k=N)
    b = random.choices(range(101),k=N)
    return a,b
def add_large_arrays_numpy_randint(N):
    a = np.random.randint(0,101,N)
    b = np.random.randint(0,101,N)
    return a,b   
    
def measure_time(func,N):
    start = time.time()
    func(N)
    return time.time() - start
    
n_time=[]
p_time=[]
nrand_time=[]
n_values = np.linspace(1, 1_000_000, 100, dtype=int)

for i in n_values:
    n_time.append(measure_time(large_arrays_choice_numpy,i))
    p_time.append(measure_time(add_large_arrays_choice,i))
    nrand_time.append(measure_time(add_large_arrays_numpy_randint,i))

plt.figure(figsize=(10, 6)) # 차트 크기 -> 그래프의 가로 크기 10인치, 세로 크기 6인치
plt.xlabel('Array Size') # x축 레이블
plt.ylabel('Time') # y축 레이블
plt.title('Performance : numpy VS numpy(randint) VS random')
plt.plot(n_values, n_time, label='numpy', color='blue')
plt.plot(n_values, p_time, label='random', color='red')
plt.plot(n_values, nrand_time, label='randint', color='green')
plt.legend() # 범례를 추가
plt.grid(True) # 그래프에 그리드 ON
plt.show()
```
